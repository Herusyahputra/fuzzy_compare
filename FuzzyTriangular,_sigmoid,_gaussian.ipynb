```python
# Example Section
import numpy as np
import matplotlib.pyplot as plt
import time
import pandas as pd

# Fuzzy Triangular Function
def fuzzy_triangular(x, a, b, c):
    return np.where((x >= a) & (x <= b), (x - a) / (b - a),
                    np.where((x > b) & (x <= c), (c - x) / (c - b), 0))

# Fuzzy Sigmoid Function
def fuzzy_sigmoid(x, a=1, b=1):
    return 1 / (1 + np.exp(-a * (x - b)))

# Fuzzy Gaussian Function
def fuzzy_gaussian(x, peak, sigma):
    return np.exp(-0.5 * ((x - peak) / sigma) ** 2)

# Create a simple classification problem with ground truth values
x_data = x_data = np.linspace(0, 10, 100)
true_positive_range = (4, 6)  # Ground truth for positive class
ground_truth = np.where((x_data >= true_positive_range[0]) & (x_data <= true_positive_range[1]), 1, 0)

# Define parameters for the membership functions
triangular_a = 2
triangular_b = 5
triangular_c = 8

sigmoid_a = 1
sigmoid_b = 5

gaussian_peak = 5.0
gaussian_sigma = 2.0

# Compute the membership values for each function
triangular_values = fuzzy_triangular(x_data, triangular_a, triangular_b, triangular_c)
sigmoid_values = fuzzy_sigmoid(x_data, sigmoid_a, sigmoid_b)
gaussian_values = fuzzy_gaussian(x_data, gaussian_peak, gaussian_sigma)

# Threshold the membership values to make binary predictions
triangular_predictions = np.where(triangular_values >= 0.5, 1, 0)
sigmoid_predictions = np.where(sigmoid_values >= 0.5, 1, 0)
gaussian_predictions = np.where(gaussian_values >= 0.5, 1, 0)

# Calculate accuracy and precision
def calculate_accuracy_precision(ground_truth, predictions):
    true_positive = np.sum(np.logical_and(ground_truth == 1, predictions == 1))
    false_positive = np.sum(np.logical_and(ground_truth == 0, predictions == 1))
    true_negative = np.sum(np.logical_and(ground_truth == 0, predictions == 0))

    accuracy = (true_positive + true_negative) / len(ground_truth)

    if true_positive + false_positive == 0:
        precision = 0.0
    else:
        precision = true_positive / (true_positive + false_positive)

    return accuracy, precision

triangular_accuracy, triangular_precision = calculate_accuracy_precision(ground_truth, triangular_predictions)
sigmoid_accuracy, sigmoid_precision = calculate_accuracy_precision(ground_truth, sigmoid_predictions)
gaussian_accuracy, gaussian_precision = calculate_accuracy_precision(ground_truth, gaussian_predictions)

# Print the results
print("Accuracy and Precision:")
print(f"Fuzzy Triangular - Accuracy: {triangular_accuracy:.4f}, Precision: {triangular_precision:.4f}")
print(f"Fuzzy Sigmoid - Accuracy: {sigmoid_accuracy:.4f}, Precision: {sigmoid_precision:.4f}")
print(f"Fuzzy Gaussian - Accuracy: {gaussian_accuracy:.4f}, Precision: {gaussian_precision:.4f}")

# Plot the membership functions (optional)
plt.figure(figsize=(10, 5))
plt.plot(x_data, triangular_values, label='Fuzzy Triangular', color='blue')
plt.plot(x_data, sigmoid_values, label='Fuzzy Sigmoid', color='green')
plt.plot(x_data, gaussian_values, label='Fuzzy Gaussian', color='red')
plt.axhline(0.5, color='black', linestyle='--', label='Threshold')
plt.title("Fuzzy Membership Functions")
plt.xlabel("Input")
plt.ylabel("Membership Value")
plt.grid(True)
plt.legend()
plt.show()

# Measure the running time for each function
start_time = time.time()
_ = fuzzy_triangular(x_data, triangular_a, triangular_b, triangular_c)
triangular_time = time.time() - start_time

start_time = time.time()
_ = fuzzy_sigmoid(x_data, sigmoid_a, sigmoid_b)
sigmoid_time = time.time() - start_time

start_time = time.time()
_ = fuzzy_gaussian(x_data, gaussian_peak, gaussian_sigma)
gaussian_time = time.time() - start_time

print(f"Fuzzy Triangular Running Time: {triangular_time:.6f} seconds")
print(f"Fuzzy Sigmoid Running Time: {sigmoid_time:.6f} seconds")
print(f"Fuzzy Gaussian Running Time: {gaussian_time:.6f} seconds")

```
